```python
import pandas as pd
import urllib.request
import zipfile
import os
```

```python
###TAKES A LONG TIME###
# Download zip files and unzip, read and append to data frame
# month loop within year loop, 2016 to 2018 pulled

years = ['2016','2017','2018','2019']
months = ['01','02','03','04','05','06','07','08','09','10','11','12']

###test data
#years = ['2017']
#months = ['03','04']

bike_df = pd.DataFrame()
failed_list = []


for y in years:
    year = y
    
    for m in months:
        
        #inconsistant url for 201708 fix
        if y == '2017' and m == '08':
            month = '08%20'
        else:
            month = f'{m}-'
        
        try:

            
            ###Directories and file names###
            url = f'https://s3.amazonaws.com/tripdata/JC-{year}{month}citibike-tripdata.csv.zip'
            zip_name = f'./JC-{year}{month}citibike-tripdata.csv.zip'
            destination = f'./csv_files'
            csv_name = f'./JC-{year}{month}citibike-tripdata.csv'
            
            #download from loop url
            urllib.request.urlretrieve(url,zip_name)
            print('--------------')
            print(f'{zip_name} saved')

            #unzip
            zip_ref = zipfile.ZipFile(zip_name, 'r')
            zip_list = zip_ref.namelist()    
            zip_ref.extract(zip_list[0])
            zip_ref.close()
            os.rename(zip_list[0], csv_name)
            print(f'{zip_name} unzipped')
            
            ###append to dataframe and add month starting column
            file_df = pd.read_csv(csv_name)
            file_df['month_starting'] = f'{month[0:2]}/01/{year}'
            
            #force headers to match first df
            if len(bike_df)>0:
                file_df.columns = bike_df.columns.values.tolist()
            else:
                file_df
                
            bike_df = bike_df.append(file_df)
            

            #remove zip file and temp csv
            os.remove(zip_name)
            os.remove(csv_name)

            print(f'{zip_name} complete')        

        except:
            print(f'{zip_name} error')
            failed_list.append(f'{year}{month}')

print('********table complete**********')
print (f'failed:{failed_list}')
```
